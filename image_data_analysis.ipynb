```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
cv2.__version__
```

```python
# loading images
img1 = cv2.imread("D:/class-fall2021-FAMU/MAT4906/Mat4906-ML/forest.jpg")
img2 = cv2.imread("D:/class-fall2021-FAMU/MAT4906/Mat4906-ML/people-walking.jpg")
print('Showing size image 1',img1.shape)
print('Showing size image 2',img2.shape)
img1_reshaped = cv2.resize(img1,(300,300),interpolation=cv2.INTER_AREA)
img2_reshaped = cv2.resize(img2,(300,300),interpolation=cv2.INTER_AREA)
print('showing size image 2 reshaped',img2_reshaped.shape)
```

```python
#showing the images
plt.figure(figsize=(10,10))
plt.subplot(221)
plt.imshow(img1_reshaped)
plt.subplot(222)
plt.imshow(img2_reshaped)
```

```python
# Printing RGB chanels
plt.figure(figsize=(10,10))
plt.subplot(131)
plt.imshow(img1_reshaped[:,:,0])
plt.gca().set_title('Red Chanel')
plt.subplot(132)
plt.imshow(img1_reshaped[:,:,1])
plt.gca().set_title('Green Chanel')
plt.subplot(133)
plt.imshow(img1_reshaped[:,:,2])
plt.gca().set_title('Blue Chanel')
```

```python
dst = cv2.addWeighted(img1_reshaped,0.2,img2_reshaped,0.5,0.1)
plt.figure(figsize=(10,10))
plt.imshow(dst)
```

```python
dst_subtraction = cv2.subtract(img1_reshaped,img2_reshaped)
plt.figure(figsize=(10,10))
plt.imshow(dst_subtraction)
```

```python
img_cat = cv2.imread("D:/class-fall2021-FAMU/MAT4906/Mat4906-ML/cat.jpg")
negative = 255 - img_cat
plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(img_cat)
plt.gca().set_title('Original image')
plt.axis('off')
plt.subplot(122)
plt.imshow(negative)
plt.gca().set_title('Negative image')
plt.axis('off')
```

```python
img_blur = cv2.blur(img_cat,(40,40))
img_filter = cv2.boxFilter(img_cat,-5,(30,30),normalize=True)
img_Gaussianfilter = cv2.GaussianBlur(img_cat,(41,41),cv2.BORDER_DEFAULT)

plt.figure(figsize=(20,20))
plt.subplot(141)
plt.imshow(img_cat)
plt.gca().set_title('Original image')
plt.axis('off')
plt.subplot(142)
plt.imshow(img_blur)
plt.gca().set_title('blur image with blur method')
plt.axis('off')
plt.subplot(143)
plt.imshow(img_filter)
plt.gca().set_title('blur image with boxFilter method')
plt.axis('off')
plt.subplot(144)
plt.imshow(img_Gaussianfilter)
plt.gca().set_title('blur image with GaussianBlur method')
plt.axis('off')
```

```python


img_median = cv2.medianBlur(img_cat,31)
plt.figure(figsize=(10,10))
plt.subplot(121)
plt.imshow(img_cat)
plt.gca().set_title('Original image')
plt.axis('off')
plt.subplot(122)
plt.imshow(img_median)
plt.gca().set_title('blur image with medianBlur method')
plt.axis('off')

```

```python

img_cat_cvt = cv2.cvtColor(img_cat, cv2.cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,10))
plt.imshow(img_cat_cvt)

```

```python
img_cat_cvt_GRAY = cv2.cvtColor(img_cat_cvt, cv2.cv2.COLOR_RGB2GRAY)
plt.figure(figsize=(10,10))
plt.imshow(img_cat_cvt_GRAY)
```

```python

img_cat_cvt_HSV = cv2.cvtColor(img_cat_cvt, cv2.cv2.COLOR_RGB2HSV)
plt.figure(figsize=(10,10))
plt.imshow(img_cat_cvt_HSV)

```

```python
print(img_cat_cvt[800,1000,:])
upper_limit = (13,119,119)
lower_limit = (61,191,191)
mask = cv2.inRange(img_cat_cvt,upper_limit,lower_limit)
result = cv2.bitwise_and(img_cat_cvt,img_cat_cvt,mask=mask)
plt.figure(figsize=(20,20))
plt.subplot(121)
plt.imshow(mask, cmap="gray")
plt.subplot(122)
plt.imshow(result)
```
